# NVIDIA Web-based Data Visualization

First, mounting the movies.csv data, which was downloaded into Google Drive

```python
import pandas as pd
df = pd.read_csv('movies.csv')
df.head()
```

## a. Creating Scatterplots



Feature 10 (Wins and nominations) vs. Feature 6 (IMDb Rating)

```python
import matplotlib.pyplot as plt

# Filter data for good and bad ratings
good_rating = df[df['IsGoodRating'] == 1]
bad_rating = df[df['IsGoodRating'] == 0]

# Create the scatter plot
plt.figure(figsize=(10, 6))

# Plot good rating instances as blue crosses
plt.scatter(good_rating['imdbRating'], good_rating['WinsNoms'],
            color='blue', marker='x', label='Good Rating')

# Plot bad rating instances as red circles
plt.scatter(bad_rating['imdbRating'], bad_rating['WinsNoms'],
            color='red', marker='o', label='Bad Rating')

# Add title and labels
plt.title('Wins+Nominations vs. IMDb Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Wins+Noms')

# Add legend to the top right corner
plt.legend(loc='upper right')

# Display the plot
plt.grid(True)
plt.show()
```

Feature 8 (Budget) vs. Features 6 (IMDb Rating)

```python
# Filter data for good and bad ratings
good_rating = df[df['IsGoodRating'] == 1]
bad_rating = df[df['IsGoodRating'] == 0]

# Create the scatter plot
plt.figure(figsize=(10, 6))

# Plot good rating instances as blue crosses
plt.scatter(good_rating['imdbRating'], good_rating['Budget'],
            color='blue', marker='x', label='Good Rating')

# Plot bad rating instances as red circles
plt.scatter(bad_rating['imdbRating'], bad_rating['Budget'],
            color='red', marker='o', label='Bad Rating')

# Add title and labels
plt.title('Budget vs. IMDb Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Budget')

# Add legend to the top right corner
plt.legend(loc='upper right')

# Display the plot
plt.grid(True)
plt.show()
```

Judging by both scatterplots we created, the first scatterplot, using Wins and Nominations vs IMDb Rating does a better job at separating the classes. This is because there is a much clearer separation data points denoted with Good Ratings and data points with Bad Ratings.

## b. Scaling Symbol sizes

```python
import matplotlib.pyplot as plt

# Filter data for good and bad ratings
good_rating = df[df['IsGoodRating'] == 1]
bad_rating = df[df['IsGoodRating'] == 0]

scaling_factor = 10

# Create the scatter plot
plt.figure(figsize=(12, 7))

# Plot good rating instances as blue crosses, with size scaled by WinsNoms
plt.scatter(good_rating['imdbRating'], good_rating['imdbVotes'],
            s=good_rating['WinsNoms'] * scaling_factor, # Scale size by WinsNoms
            color='blue', marker='x', label='Good Rating',
            alpha=0.6) # Add some transparency for overlapping points

# Plot bad rating instances as red circles, with size scaled by WinsNoms
plt.scatter(bad_rating['imdbRating'], bad_rating['imdbVotes'],
            s=bad_rating['WinsNoms'] * scaling_factor, # Scale size by WinsNoms
            color='red', marker='o', label='Bad Rating',
            alpha=0.6) # Add some transparency for overlapping points

# Add title and labels
plt.title('Votes vs. IMDb Rating sized by Wins+Nominations')
plt.xlabel('IMDb Rating')
plt.ylabel('IMDb Votes')

# Add legend
plt.legend(loc='upper right')

# Add grid for readability
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Filter data for good and bad ratings
good_rating = df[df['IsGoodRating'] == 1].copy()
bad_rating  = df[df['IsGoodRating'] == 0].copy()


good_rating['WinsNoms'] = np.sqrt(good_rating['WinsNoms'])
bad_rating['WinsNoms']  = np.sqrt(bad_rating['WinsNoms'])


# Create the scatter plot
plt.figure(figsize=(10, 6))

# Plot good rating instances as blue crosses
plt.scatter(good_rating['imdbRating'], good_rating['WinsNoms'],
            color='blue', marker='x', label='Good Rating')

# Plot bad rating instances as red circles
plt.scatter(bad_rating['imdbRating'], bad_rating['WinsNoms'],
            color='red', marker='o', label='Bad Rating')

# Add title and labels
plt.title('Wins+Nominations(sqrt) vs. IMDb Rating')
plt.xlabel('IMDb Rating')
plt.ylabel('Wins+Noms(sqrt)')

# Add legend to the top right corner
plt.legend(loc='upper right')

# Display the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Filter data for good and bad ratings
good_rating = df[df['IsGoodRating'] == 1].copy()
bad_rating  = df[df['IsGoodRating'] == 0].copy()

good_rating['imdbRating'] = np.log(good_rating['imdbRating'])
bad_rating['imdbRating']  = np.log(bad_rating['imdbRating'])



plt.figure(figsize=(10, 6))

plt.scatter(good_rating['imdbRating'], good_rating['WinsNoms'],
            color='blue', marker='x', label='Good Rating')


plt.scatter(bad_rating['imdbRating'], bad_rating['WinsNoms'],
            color='red', marker='o', label='Bad Rating')

# Add title and labels
plt.title('Wins+Nominations vs. IMDb Rating(log)')
plt.xlabel('IMDb Rating(log)')
plt.ylabel('Wins+Noms')

# Add legend to the top right corner
plt.legend(loc='upper right')

# Display the plot
plt.grid(True)
plt.show()
```

We would use square root or log visualization when the magnitude of some of our data values is wildly diffrent. This helps to squeeze the data down into a more easy to visualize manner. Example would be computer improvement. We use log when showing this because computer increases exponentialy each year and eirly years would be flattened out to a straight line if log axis were not used.
